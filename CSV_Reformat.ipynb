```python
import pandas as pd
import numpy as np
import pycountry
```

## This reformats the climate data (carbon) csv

```python
climate_csv = "climate_data.csv"

climate_df = pd.read_csv(climate_csv)

country_csv = "countries_geo.csv"

country_df = pd.read_csv(country_csv)

climate_df.head()
```

```python
climate_df = climate_df.drop(columns=['total_co2_fossil_cement', 'emissions_solid_fuel', 'emissions_liquid_fuel',
                           'emissions_gas_fuel', 'emissions_cement', 'emissions_gas_flaring', 'emissions_bunker_fuel'])

climate_df['country'] = climate_df['country'].str.title()

climate_df.drop(climate_df[climate_df.year < 1995].index, inplace=True)

climate_df.head()
```

```python
df_p = climate_df.pivot(index='country', columns='year', values = 'per_capita_co2_emissions')

df_p.head()
```

```python
country_std = []

for country, row in df_p.iterrows():
    print(country)
    try:
        result = pycountry.countries.search_fuzzy(country)
    except LookupError:
        result = 0
    if result != 0:
        string = str(result[0])
        string2 = string.split("'")
        final = string2[5]
    else:
        final = np.nan
    
    country_std.append(final)

df_p['country_std'] = country_std
```

```python
df_p.head()
```

```python
#fixing some faults in the Pycountry program that I can't fix automatically so I'm doing it by hand.
#Specifically PyCountry thinks that Niger is part of Nigeria (and is wrong)
#And while it normally lists island protectorates and overseas territories separately (e.g. Guernsey is Guernesy),
#For some of the French overseas protectorates they list them as part of France.
#These errors mean I can't use the standard country name as a Primary Key later so I'm fixing them. 

for country, row in df_p.iterrows():
    if country == ("Niger"):
        print("Niger")
        df_p.loc[country, 'country_std'] = "Niger"
    elif country == "Guadeloupe":
        print("Guadeloupe")
        df_p.loc[country, 'country_std'] = "Guadeloupe"
    elif country == "Mayotte":
        df_p.loc[country, 'country_std'] = "Mayotte"
        print("Mayotte")
    elif country == "Curacao":
        df_p.loc[country, 'country_std'] = "Curacao"
        print("Curacao")
    elif country == "Kosovo":
        df_p.loc[country, 'country_std'] = "Kosovo"
        print("Kosovo")
```

```python
df_p = df_p.dropna(subset = ['country_std'])
df_p = df_p.set_index('country_std')
df_p.head()
```

```python
len(df_p)
```

```python
df = df_p.merge(country_df, how='inner', on='country_std')

df.head()
```
